```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-5,5,0.5)
y1=2*x**2
y2=5*x+30
y3=4*x**2+10

plt.plot(x,y1)
plt.show()
plt.plot(x,y2)
plt.show()
plt.plot(x,y3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-5,5,0.5)
y1=2*x**2
y2=5*x+30
y3=4*x**2+10

plt.subplot(1,3,1)
plt.plot(x,y1)
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.subplot(1,3,3)
plt.plot(x,y3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#함수설정
x=np.linspace(-5,5,100) # -5부터 5까지 100칸으로 나누기
y1=x**3
y2=10*x**2-2

#그래프그리기
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-5,5,100)
y1=x**3
y2=10*x**2-2

plt.plot(x,y1)
plt.plot(x,y2)

plt.xlim(-1,1)
plt.ylim(-3,3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3

plt.plot(x,y1, color='r', linestyle='--', marker='o')
plt.plot(x,y2, color='b', linestyle='-', marker='s')
plt.plot(x,y3, color='g', linestyle=':', marker='*')
plt.plot(x,y4, '^k-.')#'색선마커'
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,5,1)
y1=x

#그림크기
#plt.figure(figsize=(5,5))
#화질
#plt.figure(dpi=200)
#그래프
plt.plot(x,y1, color='r', linestyle='--', marker='o', label='function')

plt.title('Graph Title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best')
plt.grid()
plt.show()
```

2. 산점도

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,5,1)
y1=x

plt.scatter(x,y1, color='r', marker='o',s=100*x)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

city=['Seoul','Incheon', 'Daejeon','Daegu', 'Ulsan', 'Busan','Gwangju']

lat=[37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]

lon=[126.97, 126.70,127.38, 128.60, 129.31, 129.07, 126.85]

pop_den=[16154, 2751, 2839, 2790, 1099, 4454, 2995]

size=np.array(pop_den)*0.2

colors=['r','g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(lon,lat, s=size, c=colors, alpha=0.5)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Population of City')

for x, y, name in zip(lon, lat, city):
    plt.text(x,y,name)

plt.ylim(34, 38)
plt.show()
```

3. 막대그래프

```python
import matplotlib.pyplot as plt
import numpy as np

member=['A', 'B', 'C', 'D']
data1=[27, 35, 40, 33]
data2=[30, 38, 42, 37]

plt.bar(range(len(member)), data1 )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

member=['A', 'B', 'C', 'D']
data1=[27, 35, 40, 33]
data2=[30, 38, 42, 37]

index=np.arange(len(member))

plt.bar(index, data1, tick_label=member,width=0.5, color='r', edgecolor='k' )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

member=['A', 'B', 'C', 'D']
data1=[27, 35, 40, 33]
data2=[30, 38, 42, 37]

index=np.arange(len(member))
barwidth=0.3

plt.bar(index, data1,width=barwidth, color='r', edgecolor='k' )
plt.bar(index+barwidth, data1, tick_label=member,width=barwidth, color='y', edgecolor='k' )
plt.xticks(index+0.15, member)
plt.show()
```
