```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f=open('people.csv', encoding='cp949')
data=csv.reader(f)

result=[]
for row in data:
    if '종로구 사직동' in row[0]:
        for i in row[3:104]:
            result.append(int(i))

plt.bar(range(0,101),result, color='r', label='total')
plt.xlabel('age')
plt.ylabel('population')
plt.legend()
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f=open('people.csv', encoding='cp949')
data=csv.reader(f)

m_result=[]
f_result=[]
for row in data:
    if '분당구 판교동' in row[0]:#구로구 가리봉동도 입력해봅시다. (난쏘공)
        for i in row[106:207]:
            m_result.append(int(i))
        for i in row[209:310]:
            f_result.append(int(i))

plt.plot(range(0,101),m_result, color='b', label='male')
plt.plot(range(0,101),f_result, color='r', label='female')
plt.xlabel('age')
plt.ylabel('population')
plt.legend()
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f=open('anscombe_data.csv')
data=csv.reader(f)

next(data)

x=[]
y=[]

for row in data:
    if row[1]=='4':
        x.append(float(row[2]))
        y.append(float(row[3]))


plt.scatter(x,y)
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
x_std=np.std(x)
y_std=np.std(y)

print(x_mean,y_mean, x_std, y_std)
```
